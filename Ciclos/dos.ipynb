```python
def main():
    # Solicitar al usuario un número entero
    num_entero = int(input("Por favor, ingresa un número entero: "))

    # Imprimir el triángulo rectángulo
    print("Triángulo rectángulo:")
    for i in range(1, num_entero + 1):
        print("*" * i)

if __name__ == "__main__":
    main()
```
