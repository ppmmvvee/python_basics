```python
def main():
    # Solicitar al usuario la cantidad a invertir, el interés anual y el número de años
    cantidad = float(input("Ingrese la cantidad a invertir: "))
    interes_anual = float(input("Ingrese el interés anual (en porcentaje): "))
    años = int(input("Ingrese el número de años de la inversión: "))

    # Convertir el interés anual de porcentaje a decimal
    interes_anual /= 100

    # Calcular el capital obtenido para cada año de la inversión y mostrarlo por pantalla
    for año in range(1, años + 1):
        capital = cantidad * (1 + interes_anual) ** año
        print(f"Capital obtenido en el año {año}: {capital:.2f}")

if __name__ == "__main__":
    main()
```
