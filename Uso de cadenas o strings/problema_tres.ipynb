```python
def contar_letras_nombre(nombre_usuario):
    # Contar el número de letras en el nombre del usuario
    numero_letras = len(nombre_usuario)
    # Convertir el nombre a mayúsculas
    nombre_mayusculas = nombre_usuario.upper()
    # Mostrar el mensaje por pantalla
    print(f"{nombre_mayusculas} tiene {numero_letras} letras.")

# Ejemplo de uso:
nombre = input("Ingrese su nombre: ")
contar_letras_nombre(nombre)
```
