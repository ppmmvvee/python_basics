```python
def invertir_frase(frase):
    # Invertir la frase utilizando slicing con paso -1
    frase_invertida = frase[::-1]
    # Mostrar la frase invertida por pantalla
    print("Frase invertida:", frase_invertida)

# Ejemplo de uso:
frase = input("Ingrese una frase: ")
invertir_frase(frase)
```
