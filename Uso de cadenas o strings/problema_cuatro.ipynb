```python
def obtener_numero_telefono(numero_telefono):
    # Dividir el número de teléfono en partes usando el guion como separador
    partes = numero_telefono.split("-")

    # Extraer el número de teléfono sin el prefijo ni la extensión
    numero_sin_prefijo_extension = partes[1]

    # Mostrar el número de teléfono sin el prefijo ni la extensión por pantalla
    print("Número de teléfono sin prefijo ni extensión:", numero_sin_prefijo_extension)

# Ejemplo de uso:
numero_telefono = "+34-913724710-56"
obtener_numero_telefono(numero_telefono)
```
