```python
def decimal_a_binario(numero_decimal):
    if numero_decimal == 0:
        return '0'

    binario = ''
    while numero_decimal > 0:
        residuo = numero_decimal % 2
        binario = str(residuo) + binario
        numero_decimal = numero_decimal // 2

    return binario

# Ejemplo de uso:
numero_decimal = int(input("Ingrese un número decimal: "))
binario = decimal_a_binario(numero_decimal)
print("El número binario equivalente es:", binario)


def binario_a_decimal(numero_binario):
    decimal = 0
    longitud = len(numero_binario)

    for i in range(longitud):
        bit = int(numero_binario[longitud - i - 1])
        decimal += bit * (2 ** i)

    return decimal

# Ejemplo de uso:
numero_binario = input("Ingrese un número binario: ")
decimal = binario_a_decimal(numero_binario)
print("El número decimal equivalente es:", decimal)
```
