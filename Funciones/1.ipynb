```python
def calcular_total_factura(cantidad_sin_iva, porcentaje_iva=21):
    """
    Calcula el total de una factura tras aplicarle el IVA.

    Parámetros:
    - cantidad_sin_iva: La cantidad sin IVA.
    - porcentaje_iva: El porcentaje de IVA a aplicar (por defecto es 21%).

    Devuelve:
    - El total de la factura después de aplicarle el IVA.
    """
    total = cantidad_sin_iva + (cantidad_sin_iva * porcentaje_iva / 100)
    return total

# Ejemplos de uso:
cantidad = float(input("Ingrese la cantidad sin IVA: "))
porcentaje_iva = float(input("Ingrese el porcentaje de IVA (si no ingresa, se aplicará el 21%): ") or 21)

total_factura = calcular_total_factura(cantidad, porcentaje_iva)
print("El total de la factura es:", total_factura)
```
