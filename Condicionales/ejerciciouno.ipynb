```python
def es_mayor_de_edad(edad):
    if edad >= 18:
        print("Es mayor de edad.")
    else:
        print("No es mayor de edad.")

# Ejemplo de uso:
try:
    edad = int(input("Ingrese su edad: "))
    es_mayor_de_edad(edad)
except ValueError:
    print("Error: Por favor, ingrese un n√∫mero entero para la edad.")
```
