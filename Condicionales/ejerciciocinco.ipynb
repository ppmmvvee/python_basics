```python
def calcular_dinero(puntuacion):
    if puntuacion == 0.0:
        nivel = "Inaceptable"
    elif puntuacion == 0.4:
        nivel = "Aceptable"
    elif puntuacion >= 0.6:
        nivel = "Meritorio"
    else:
        return "Puntuación inválida", 0

    dinero_conseguido = puntuacion * 2400
    return nivel, dinero_conseguido

def main():
    try:
        puntuacion = float(input("Por favor, ingrese la puntuación del empleado: "))
        if puntuacion < 0.0:
            print("La puntuación no puede ser menor que 0.0")
            return
        nivel, dinero = calcular_dinero(puntuacion)
        print(f"El nivel de rendimiento del empleado es: {nivel}")
        print(f"El empleado recibirá {dinero}€ como beneficio.")
    except ValueError:
        print("Error: Por favor, ingrese un valor numérico válido.")

if __name__ == "__main__":
    main()
```
