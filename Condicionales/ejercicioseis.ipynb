```python
def calcular_precio_entrada(edad):
    if edad < 4:
        precio = 0
    elif 4 <= edad <= 18:
        precio = 5
    else:
        precio = 10
    return precio

def main():
    try:
        edad = int(input("Por favor, ingresa la edad del cliente: "))
        if edad < 0:
            print("La edad no puede ser negativa.")
            return
        precio_entrada = calcular_precio_entrada(edad)
        print(f"El precio de la entrada para un cliente de {edad} años es: {precio_entrada}€")
    except ValueError:
        print("Error: Por favor, ingresa una edad válida.")

if __name__ == "__main__":
    main()
```
