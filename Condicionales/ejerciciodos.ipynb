```python
def verificar_contrasena(contrasena_guardada):
    # Pedir al usuario que ingrese la contraseña
    contrasena_usuario = input("Por favor, ingrese la contraseña: ")

    # Convertir ambas contraseñas a minúsculas para compararlas sin importar mayúsculas y minúsculas
    contrasena_guardada = contrasena_guardada.lower()
    contrasena_usuario = contrasena_usuario.lower()

    # Verificar si las contraseñas coinciden
    if contrasena_guardada == contrasena_usuario:
        print("¡La contraseña es correcta!")
    else:
        print("La contraseña es incorrecta.")


# Ejemplo de uso de la función
contrasena_guardada = "MiContrasenaSegura123"
verificar_contrasena(contrasena_guardada)
```
