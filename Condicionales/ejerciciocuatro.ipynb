```python
def determinar_grupo(nombre, sexo):
    nombre = nombre.lower()

    # Verificar el sexo y el nombre para determinar el grupo
    if sexo.lower() == 'femenino':
        if nombre < 'm':
            grupo = 'A'
        else:
            grupo = 'B'
    else:  # Si el sexo no es femenino, asumimos que es masculino
        if nombre >= 'n':
            grupo = 'A'
        else:
            grupo = 'B'


    return grupo


# Pedir al usuario que ingrese su nombre y sexo
nombre = input("Ingrese su nombre: ")
sexo = input("Ingrese su sexo (Femenino o Masculino): ")

# Determinar el grupo al que pertenece el usuario
grupo = determinar_grupo(nombre, sexo)

# Mostrar por pantalla el grupo al que pertenece el usuario
print("Usted pertenece al grupo:", grupo)
```
