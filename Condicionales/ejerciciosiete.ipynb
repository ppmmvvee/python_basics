```python
def main():
    # Mostramos las opciones al usuario
    print("Bienvenido a Bella Napoli")
    print("¿Qué tipo de pizza te gustaría ordenar?")
    print("1. Pizza vegetariana")
    print("2. Pizza no vegetariana")

    # Solicitamos al usuario que elija una opción
    tipo_pizza = input("Ingresa el número correspondiente a tu elección: ")

    if tipo_pizza == "1":
        print("Ingredientes disponibles para la pizza vegetariana:")
        print("1. Pimiento")
        print("2. Tofu")
        ingrediente = input("Elige un ingrediente para tu pizza: ")
        if ingrediente == "1":
            ingrediente_elegido = "Pimiento"
        elif ingrediente == "2":
            ingrediente_elegido = "Tofu"
        else:
            print("Opción inválida")
            return
        tipo = "vegetariana"
    elif tipo_pizza == "2":
        print("Ingredientes disponibles para la pizza no vegetariana:")
        print("1. Peperoni")
        print("2. Jamón")
        print("3. Salmón")
        ingrediente = input("Elige un ingrediente para tu pizza: ")
        if ingrediente == "1":
            ingrediente_elegido = "Peperoni"
        elif ingrediente == "2":
            ingrediente_elegido = "Jamón"
        elif ingrediente == "3":
            ingrediente_elegido = "Salmón"
        else:
            print("Opción inválida")
            return
        tipo = "no vegetariana"
    else:
        print("Opción inválida")
        return

    # Mostramos la pizza elegida y sus ingredientes
    print(f"Has elegido una pizza {tipo} con mozzarella, tomate y {ingrediente_elegido}.")

if __name__ == "__main__":
    main()
```
